```python
df=pd.read_csv("diabetes - diabetes.csv")
```

```python
df.head()
```

```python
import pandas as pd




# Remove special characters
df = df.replace('#', '', regex=True)

# Print the modified DataFrame
print(df)
```

```python
feature_cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
x=df[feature_cols] #features
y=df.Outcome #target variable
```

```python
# split dataset into training and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
clf=DecisionTreeClassifier(random_state=1)
# train decision tree classifier
clf=clf.fit(x_train,y_train)
y_Pred=clf.predict(x_test)
```

```python
from sklearn.tree import export_graphviz
from sklearn import tree
from io import StringIO
from IPython.display import Image
import pydotplus

# Generate DOT data
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names=feature_cols, class_names=['0', '1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

# Save the decision tree visualization as PNG
graph.write_png('diabetes.png')

# Display the decision tree image
Image(graph.create_png())
```

```python
clf=DecisionTreeClassifier(random_state=1,max_depth=3, min_samples_leaf=10)
# train decision tree classifier
clf=clf.fit(x_train,y_train)
```

```python


x_train_pred=clf.predict(x_test)
y_train_pred=clf.predict(x_train)
```

```python
# model accuracy how often is the classifier correct
print('traininng accuracy:',metrics.accuracy_score(y_train,y_train_pred))
print("testing accuracy:",metrics.accuracy_score(y_test,y_Pred))
```
